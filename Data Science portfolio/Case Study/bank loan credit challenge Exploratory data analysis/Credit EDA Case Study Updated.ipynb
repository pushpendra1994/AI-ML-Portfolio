## Binning the Variables
#### Binning the DAYS_BIRTH variable

```python
df_binned = pd.DataFrame()
df_binned['BIRTH_YEAR'] = df_app['DAYS_BIRTH'].apply(lambda x : int(x/365))  
# Here we make new column as BIRH_YEAR and change the days onto year wise
df_binned['BIRTH_YEAR'].head()
```

```python
# Defining binning as 0-18, 18-25, 25-35, 35-55, 55-75, 75-100
bins = [0, 18, 25,35, 60, 75, 100]
# Names the Age Groups
group_names = ['Teen', 'Young', 'Adult', 'Mid Adult', 'Senior Citizen', 'Old']
df_binned['Age Group'] = pd.cut(df_binned['BIRTH_YEAR'], bins, labels = group_names)
df_binned['Age Group'].value_counts()
```

#### Binning the DAYS_ID_PUBLISH variable


```python
# Here we make new column as YEAR_ID_PUBLISH and change the days onto year wise
bins = [0, 5, 10, 15, 20]
df_binned['YEAR_ID_PUBLISH'] = df_app['DAYS_ID_PUBLISH'].apply(lambda x: int(x/365)) 
df_binned['ID_PUBLISH_BINNED'] = pd.cut(df_binned['YEAR_ID_PUBLISH'], bins)
df_binned['ID_PUBLISH_BINNED'].value_counts()
```

#### Binning the DAYS_EMPLOYED variable


```python
df_binned['YEARS_EMPLOYED'] = df_app['DAYS_EMPLOYED'].apply(lambda x: int(x/365))
bins = [0, 2, 4, 6, 8, 10, 12, 15, 16]
df_binned['YEARS_EMPLOYED_BINNED'] = pd.cut(df_binned['YEARS_EMPLOYED'], bins)  # Here we make new column as DAYS_EMPLOYED and change the days onto year wise
df_binned['YEARS_EMPLOYED_BINNED'].value_counts()
```

## Correlation

#### Correlation for Target 0 (Not Default) for dataframe

```python
corr = df0.corr()
corr = corr.where(np.triu(np.ones(corr.shape), k = 1).astype(np.bool))
corrdf0 = corr.unstack().reset_index()
corrdf0.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf0.dropna(subset = ['Correlation'], inplace = True)
# SInce we see Correlation as an absolute value, we are converting it into absolute value
corrdf0['Correlation'] = round(corrdf0['Correlation'], 2).abs()
corrdf0 = corrdf0.sort_values(by = 'Correlation', ascending = False)
corrdf0.head(10)
```

#### Correlation for Target 1 (Default) for dataframe

```python
corr = df1.corr()
corr = corr.where(np.triu(np.ones(corr.shape), k = 1).astype(np.bool))
corrdf1 = corr.unstack().reset_index()
corrdf1.columns = ['VAR1', 'VAR2', 'Correlation']
corrdf1.dropna(subset = ['Correlation'], inplace = True)
# SInce we see Correlation as an absolute value, we are converting it into absolute value
corrdf1['Correlation'] = round(corrdf1['Correlation'], 2).abs()
corrdf1 = corrdf1.sort_values(by = 'Correlation', ascending = False)
corrdf1.head(10)
```

### As we can see from above correlation from both Target 1 and 0, There are number of observation we can point out .


```python
# 1) In both the cases AMT_GOODS_PRICE and AMT_CREDIT highly correlated.
# 2) Credit amount is inversely proportional to the date of birth, which means Credit amount is higher for low age and vice-versa.
# 3) Credit amount is inversely proportional to the number of children client have, means Credit amount is higher for less children count client have and vice-versa.
# 4) Income amount is inversely proportional to the number of children client have, means more income for less children client have and vice-versa.
# 5) less children client have in densely populated area
# 6) Credit amount is higher to densely populated area.
# 7) The income is also higher in densely populated area.
```
